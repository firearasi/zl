```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
data=pd.read_csv("yxz.txt", sep='\t')
```

```python
data=data.values
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ys=data[:,0]
xs=data[:,1]
zs=data[:,2]
ax.scatter(xs,ys,zs)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6)
# Fitting with inputs
kmeans = kmeans.fit(data)
# Predicting the clusters
labels = kmeans.predict(data)
# Getting the cluster centers
C = kmeans.cluster_centers_
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=labels)
#ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
```

```python
from sklearn.cluster import AffinityPropagation
affinity=AffinityPropagation(damping=0.6)
model = affinity.fit(data)
# Predicting the clusters
labels = affinity.predict(data)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=labels)
```

```python

affinity=AffinityPropagation(damping=0.9)
model = affinity.fit(data)
# Predicting the clusters
labels = model.predict(data)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=labels)
```

```python
algo=cluster.DBSCAN(eps=3, min_samples=2)
model = algo.fit(data)
# Predicting the clusters
labels = model.labels_
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], c=labels)
```
